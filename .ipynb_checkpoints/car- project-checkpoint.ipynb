```python
x_data = df[['horsepower','curb-weight','highway-mpg','engine-size','engine-location','make','wheel-base','bore','diesel','city-mpg']]
y_data = df[['price']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
lm.fit(x_train,y_train)
yhat= lm.predict(x_data)
yhat[0:5]
lm.score(x_test,y_test)
```

```python
Rsque =[]
order=[1,2,3,4,5,6]
from sklearn.preprocessing import PolynomialFeatures
for n in order:
    pr2= PolynomialFeatures(degree=n)
x_train_pr2= pr2.fit_transform(x_train)
x_test_pr2 = pr2.fit_transform(x_test)
lm.fit(x_train,y_train)
Rsque.append(lm.score(x_test,y_test))
np.mean(Rsque)
```

```python
from sklearn.linear_model import Ridge
ridgemodel = Ridge(alpha=0.1)
ridgemodel.fit(x_train ,y_train)
yhat_ridge=ridgemodel.predict(x_test)
print(yhat_ridge[0:5])
score_ridge=ridgemodel.score(x_test,y_test)
print(np.mean(score_ridge))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [{'alpha':[0.001,0.1,1,10,100,1000,10000,100000]}]
RR=Ridge()
grid1 = GridSearchCV(RR,parameters,cv=4)
grid2 = GridSearchCV(RR,parameters,cv=4)
grid1.fit(x_train,y_train)
grid2.fit(x_test,y_test)
print('grid1_bestestimator',grid1.best_estimator_)
print('grid2_bestestimator',grid2.best_estimator_)
scores1= grid1.cv_results_
scores2= grid2.cv_results_
print('mean_test_score1',scores1['mean_test_score'])
print('mean_test_score2',scores2['mean_test_score'])
```
